```python
# import required packages here
import pandas as pd  #pandas to read data
import collections   # 
from ast import literal_eval  #
from itertools import permutations #
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# load the basket.csv file into a pandas DataFrame (df)
df = pd.read_csv(r'D:\Documents\MS\UCD\MIS41270-Data Management & Mining\Assignment1-Association Rules\assignment_1\data\basket.csv') 

# Use df.shape to see the number of rows and columns in the df
df.shape
```

```python
# Use df.head() to see the first 8 rows of the df
df.head()
```

```python
# Set the index of the df to the Transaction_ID column
df.set_index('ID', inplace=True)

# Use df.head() to see the first 3 rows of the df
df.head()
```

```python
# Use the pandas describe method to get an overview of the dataset
df.describe()
```

```python
# create a list of lists, where each list contains the items in a given transaction
# instantiate an empty list
transactions = []

# iterate through each row and column to create the list
for i in range(0, len(df)): 
    transactions.append([str(df.values[i,j]) for j in range(0, len(df.columns))])

# view the first transaction in the list
transactions[:1]
```

```python
# from your list of lists, create a flattened list that contains all items purchased
flattened = [item for transaction in transactions for item in transaction]

# view the length of the list
print(len(flattened))

flattened[:20]
```

```python
# create a list of unique items from the flattened list
# use the set() method because a set by definition contains only unique items
items = list(set(flattened))

#print(items)

# remove any elements that items of interest (like empty strings or nan values)
for i in range(0, len(flattened)):
    if 'nan' in flattened: flattened.remove('nan')
if 'nan' in items: items.remove('nan')


# print out how many unique items we have
print('# of items:',len(flattened))
print('# of items:',len(items))
```

```python
# generate the itemset permutations up to 2-itemsets
rules = list(permutations(items, 2))


# print out the number of rules
print('# of rules:',len(rules))

# print some of the elements of the list of rules
print(rules[:5])
```

```python
# generate the frequency counts for each unique item in the dataset
# let's use the list that contains all items and use the Counter method
item_freq = collections.Counter(flattened)

# show the most common
item_freq.most_common()
```

```python
# one-hot encode the data and show the first 5 rows of the resulting df
encoder = TransactionEncoder().fit(transactions) # create an encoder object that is fit to our list of lists we created earlier
onehot = encoder.transform(transactions) # create encoded array

# don't forget to drop the 'nan' value
df_onehot = pd.DataFrame(onehot, columns=encoder.columns_).drop('nan', axis=1) #converting the Boolean array to Pandas dataframe & dropping 'nan' values
df_onehot.head() # first 5 rows of the resulting df
```

```python
# Generate frequent itemsets with a minimum support of 0.5%
df_itemsets = apriori(df_onehot, min_support=0.005, use_colnames=True)

# view the itemsets sorted by most frequent
df_itemsets.sort_values(by=['support'], ascending=False)
```

How many itemsets are generated for the following support thresholds?:
* 0.1%
* 0.5%
* 1%
* 10%

```python
# Generate frequent itemsets with a minimum support of 0.1%
df_itemsets1 = apriori(df_onehot, min_support=0.001, use_colnames=True)

# Generate frequent itemsets with a minimum support of 1%
df_itemsets3 = apriori(df_onehot, min_support=0.01, use_colnames=True)

# Generate frequent itemsets with a minimum support of 10%
df_itemsets4 = apriori(df_onehot, min_support=0.1, use_colnames=True)

```

```python
#Display Number of itemsets generated for each of the support threshold
print("For 0.1% support threshold: ",df_itemsets1.shape)
print("For 0.5% support threshold: ",df_itemsets.shape)
print("For 1% support threshold: ",df_itemsets3.shape)
print("For 10% support threshold: ",df_itemsets4.shape)
```

```python
# generate association rules with a confidence threshold of 10%
rules_df = association_rules(df_itemsets, metric='confidence', min_threshold=0.1)

# how many rules are generated?
rules_df.shape
```

```python
rules_df
```

```python
# print out the rules sorted by lift
print(rules_df.sort_values(by=['lift'], ascending=False))
```
